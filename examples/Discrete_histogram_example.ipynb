### This example shows how to plot discrete histograms

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import plot_utils as pu
```

```python
%matplotlib inline
```

### Discrete histogram on numerical values

```python
discrete_values = [-10, -1, 11, 3, 5, 2, 4, 0]  # define 8 discrete values
counts = [10, 20, 8, 14, 22, 15, 14, 17]   # how many times each value occurs

x = []
for j, val in enumerate(discrete_values):
    x.extend([val] * counts[j])

print(x)
```

Plot histogram; note that the numerical values are automatically sorted:

```python
pu.discrete_histogram(x);  # generate discrete histogram with default arguments
```

### Discrete histogram on categorical values

```python
discrete_values = ['cat', 'dog', 'fish', 'sheep', 'bird', 'rabbit', 'pig', 'cow']  # define 8 discrete values
counts = [10, 20, 8, 14, 22, 15, 14, 17]   # how many times each value occurs

x = []
for j, val in enumerate(discrete_values):
    x.extend([val] * counts[j])

print(x)
```

Plot histogram, with some modified argument values; note that the categories are also automatically sorted (alphabetically):

```python
fig = plt.figure(dpi=125)
ax  = plt.axes()
fig, ax, _ = pu.discrete_histogram(x, fig, ax, color='orange', alpha=0.6, rot=45);
ax.grid(lw=0.5, ls=':')
ax.set_ylabel('Number of each animals');
```
