### This example shows how to produce violin plots and "multiple histogram" plots from data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import plot_utils as pu
```

```python
%matplotlib inline
```

### Load the "iris" dataset

```python
iris = pd.read_csv('./datasets/iris.csv')
```

```python
iris.head()
```

### Produce violin plots of the four numeric columns of `iris`:

```python
pu.violin_plot(iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]);
```

### Produce multi-histogram plots of the four numeric columns of `iris`:

```python
pu.hist_multi(iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]);
```

Optionally, you can show median values:

```python
pu.hist_multi(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']],
    showmedians=True,
);
```

You can pass a numpy array (i.e., no column names), then the tick labels will be `data_0`, `data_1`, `data_2`, etc.

```python
data = np.random.randn(10, 15)  # generate 15 data sets, each of which has 10 data points
pu.violin_plot(data, vert=True);  # vert=True turns the plot by 90 degrees
```

Plot the same data with `hist_multi()`:

```python
pu.hist_multi(data, vert=True);
```

### Use other input formats

You can pass a Python dictionary, which enables different lengths and data types.

```python
d1 = {
    'cat': [0, 5, 5, 6, 6, 7, 7, 7, 7, 8, 8, 8, 9, 9, 9, 10, 15],
    'dog': np.array([1, 1, 2, 2, 2, 2, 3, 3]),
    'fish': pd.Series([7, 7, 7, 7, 8, 8, 8, 8, 9, 9, 9, 9]),
}
pu.violin_plot(d1);
```

Plot the same data with `hist_multi()`:

```python
pu.hist_multi(d1);
```

### Options to control mean/median values printed on multiple histogram plots

Default value:

```python
pu.hist_multi(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']],
);
```

Show median values:

```python
pu.hist_multi(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']],
    showmedians=True,
);
```

Turn off showing values:

```python
pu.hist_multi(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']],
    showmedians=True,
    show_vals=False,
);
```

Also show percent differences:

```python
pu.hist_multi(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']],
    vert=False,
    figsize=(8, 4),
    showmedians=True,
    show_vals=True,
    show_pct_diff=True,
);
```

You can configure which data set is considered the "baseline":

```python
pu.hist_multi(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']],
    vert=False,
    figsize=(8, 4),
    showmedians=True,
    show_vals=True,
    show_pct_diff=True,
    baseline_data_index=2,  # "2" is the index of "petal_length"
);
```
