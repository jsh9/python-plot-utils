### This notebook demonstrates how to use the `visualize_cv_scores` function

```python
import plot_utils as pu
```

Calling `visualize_cv_scores` withou any parameters generates a generic visualization:

```python
pu.visualize_cv_scores();
```

You can specify the number of folds by `n_folds`:

```python
pu.visualize_cv_scores(n_folds=3);
```

```python
pu.visualize_cv_scores(n_folds=7);
```

Specify the validation score of each cross-validation fold via `cv_scores`, as well as `metric_name`:

```python
pu.visualize_cv_scores(n_folds=4, cv_scores=[0.675, 0.702, 0.689, 0.653], metric_name='Accuracy');
```

If you have evaluated your model on the hold-out set, specify the hold-out set score via `holdout_score`:

```python
pu.visualize_cv_scores(n_folds=4, cv_scores=[0.675, 0.702, 0.689, 0.653], holdout_score=0.6745);
```

If you do not have a hold-out set at all, you can turn `no_holdout_set` to `True`:

```python
pu.visualize_cv_scores(n_folds=4, cv_scores=[0.675, 0.702, 0.689, 0.653], no_holdout_set=True);
```
